Веса могут быть присвоены объектам на основе их физических характеристик, социально-экономического влияния, и других ключевых параметров, выявленных в ходе анализа данных. Это позволит системе:

- **Объективно Оценивать Приоритет:** Веса помогут в объективной оценке и сравнении разных объектов, учитывая множество переменных.
- **Сбалансировать Ресурсы:** Присвоение весов обеспечивает более эффективное распределение ресурсов, направляя их туда, где они будут иметь наибольший эффект.
- **Учитывать Разнообразные Параметры:** Веса могут учитывать различные аспекты каждого объекта, включая его размер, возраст, использование, и множество других факторов.

```python
# Импорт необходимых библиотек
import pandas as pd

# Загрузка данных
data = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')

# Создание пустого DataFrame для сбора статистики
statistics = pd.DataFrame()

# Исследование переменных
for column in data.columns:
    # Проверка типа данных
    if pd.api.types.is_numeric_dtype(data[column]):
        # Добавление статистики для числовой переменной
        stats = data[column].describe()
        stats['var'] = data[column].var()  # Добавление дисперсии
        stats['mode'] = data[column].mode()[0]  # Добавление моды
        
    else:
        # Добавление статистики для категориальной переменной
        stats = data[column].describe()
        # Добавление самой часто встречающейся категории
        stats['top_freq'] = data[column].value_counts().values[0] 
        
    # Трансформация и объединение статистик
    stats = pd.DataFrame(stats).transpose()
    statistics = pd.concat([statistics, stats], ignore_index=False)

# Сохранение статистики в Excel-файл
statistics.to_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\statistics.xlsx')

# Вывод информации на экран
pd.set_option('display.max_columns', None)  # Отображение всех столбцов
pd.set_option('display.max_rows', None)     # Отображение всех строк
print(statistics)
```

### 2.4. Анализ Числовых Переменных:
   - **Корреляционный анализ:** Определить отношения между числовыми переменными.
   - **Визуализация корреляций:** Использовать тепловые карты или scatter plots для визуализации взаимосвязей.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Загрузка данных
file_path = 'i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx'

data = pd.read_excel(file_path)

# Выберем только числовые столбцы для корреляционного анализа
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Корреляционный анализ
corr_matrix = numeric_data.corr()

# Визуализация корреляций с использованием тепловой карты
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Тепловая карта корреляций между числовыми переменными')
plt.show()

# Основной статистический анализ
numeric_stats = numeric_data.describe().T[['mean', '50%', 'std', '25%', '75%']]
numeric_stats.columns = ['Среднее', 'Медиана', 'Стандартное отклонение', '1-й квартиль', '3-й квартиль']
print("Основные статистические показатели числовых переменных:")
print(numeric_stats)

# Анализ аномалий (выбросов) с использованием boxplot
num_vars = numeric_data.columns.tolist()
for var in num_vars:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=numeric_data[var])
    plt.title(f'Boxplot для переменной {var}')
    plt.show()
```

### 2.6. Групповой Анализ и ANOVA
   - **Поиск зависимостей:** Изучить динамику и зависимости между разными группами и категориями данных.
   - **Статистические тесты:** Применить подходящие статистические тесты для определения статистической значимости различий.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Загрузка данных
data = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')
data.columns = [col.replace(" ", "_") for col in data.columns]

# Переменные для анализа
analysis_pairs = [
    ('Материал_стен', 'Общая_площадь'),
    ('Материал_кровли', 'Количество_этажей'),
    ('Статус_МКД', 'Количество_квартир'),
    ('Статус_управления_МКД', 'Количество_подъездов')
]

# Анализ
for cat_var, num_var in analysis_pairs:
    # ANOVA анализ
    formula = f"{num_var} ~ {cat_var}"
    model = ols(formula, data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    print(f"\nANOVA for {cat_var} and {num_var}:")
    print(anova_table)
    
    # Групповой анализ
    grouped_data = data.groupby(cat_var)[num_var].describe()
    print(f"\nGrouped data for {cat_var} and {num_var}:")
    print(grouped_data)
```

   Вывод: Существует статистически значимое влияние статуса МКД на количество квартир.

### 2.7. Многомерный анализ


```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Загрузка данных
data_path = 'i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx'
data = pd.read_excel(data_path)

# Выберем только числовые столбцы для PCA
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Стандартизация данных перед PCA
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Применение PCA
pca = PCA(n_components=2)  # Для визуализации выберем первые две главные компоненты
principal_components = pca.fit_transform(scaled_data)

# Визуализация результатов PCA
plt.figure(figsize=(10, 8))
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Главная компонента 1')
plt.ylabel('Главная компонента 2')
plt.title('2D PCA Результат')
plt.grid(True)
plt.show()

# Объясненная дисперсия каждой компоненты
explained_variance = pca.explained_variance_ratio_
print(f'Объясненная дисперсия: {explained_variance}')
```

PCA (Principal Component Analysis, или метод главных компонент) – это метод уменьшения размерности данных, при котором сохраняется максимум информации. Главные компоненты, полученные методом, представляют собой новые переменные (оси), которые являются линейными комбинациями оригинальных переменных. При этом первая главная компонента объясняет максимум дисперсии в данных, вторая – следующий по величине максимум и так далее.

В данном контексте:

Главная компонента 1 объясняет 46.6% дисперсии всех числовых данных в вашем наборе данных.
Главная компонента 2 объясняет 14.8% дисперсии данных.
Понимание того, что физически представляют собой эти компоненты, может потребовать дополнительного анализа, так как они представляют собой комбинации оригинальных переменных. Исходные переменные, которые наиболее сильно влияют на каждую компоненту, могут быть определены путём анализа коэффициентов загрузки для каждой компоненты. Это может помочь лучше интерпретировать результаты.

Следующие шаги: Можно применить методы кластеризации к главным компонентам для выявления возможных групп или сегментов в данных. Также можно рассмотреть использование главных компонент как входных данных для других методов машинного обучения для предсказания интересующих вас переменных.

Для кластеризации данных с использованием главных компонент, я буду использовать метод k-средних (k-means clustering). Определение оптимального числа кластеров часто осуществляется с помощью метода "локтя", который анализирует внутрикластерное расстояние (сумму квадратов расстояний от точек до центроидов их кластеров) для различного числа кластеров и выбирает тот, начиная с которого прирост внутрикластерного расстояния становится незначительным.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Загрузка данных
data = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')

# Выбор числовых переменных для кластеризации
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Стандартизация данных
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# Определение оптимального числа кластеров с использованием метода "локтя"
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
    
# Визуализация метода "локтя"
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Метод локтя для определения числа кластеров')
plt.xlabel('Число кластеров')
plt.ylabel('WCSS')  
plt.grid(True)
plt.show()

# Кластеризация с выбранным числом кластеров (предположим, 4)
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(scaled_data)

# Добавление меток кластера к исходным данным
data['Cluster'] = clusters

# Анализ каждого кластера
for cluster in data['Cluster'].unique():
    print(f"Кластер {cluster}:")
    print(data[data['Cluster'] == cluster].describe().T[['mean', '50%', 'std', 'min', 'max']])
    
# Дополнительный анализ и визуализация данных в кластерах может включать:
# - Столбчатые диаграммы средних значений переменных по кластерам
# - Распределение категориальных переменных в каждом кластере
# - Профилирование кластеров на основе ключевых метрик и характеристик
```

### 2.8. Анализ временных рядов

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Загрузите данные
data = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')

# Убедитесь, что данные отсортированы по году постройки
data = data.sort_values('Год постройки')

# Постройте график распределения построек по годам
plt.figure(figsize=(12, 6))
sns.histplot(data['Год постройки'], bins=30, kde=True)
plt.title('Распределение количества построек по годам')
plt.xlabel('Год постройки')
plt.ylabel('Количество построек')
plt.show()

# Постройте временной ряд количества построек по годам
time_series = data['Год постройки'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(time_series.index, time_series.values, marker='o')
plt.title('Временной ряд количества построек')
plt.xlabel('Год постройки')
plt.ylabel('Количество построек')
plt.show()

# Проведите декомпозицию временного ряда для выявления трендов и сезонности
decomposed = seasonal_decompose(time_series, period=5)  # укажите период сезонности
plt.figure(figsize=(12, 8))
plt.subplot(311)
plt.plot(decomposed.trend)
plt.title('Trend')
plt.subplot(312)
plt.plot(decomposed.seasonal)
plt.title('Seasonality')
plt.subplot(313)
plt.plot(decomposed.resid)
plt.title('Residuals')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Загрузка данных
data = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')

# Подготовка данных для временного ряда по годам
time_series_data = data['Год постройки'].value_counts().sort_index()

# Визуализация данных
plt.figure(figsize=(12, 6))
plt.plot(time_series_data, marker='o', linestyle='-')
plt.title('Количество построенных зданий по годам')
plt.xlabel('Год постройки')
plt.ylabel('Количество зданий')
plt.grid(True)
plt.show()

# Проверка стационарности
result = adfuller(time_series_data)
print('ADF Test Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.6f' % (key, value))

# Автокорреляция (ACF) и частичная автокорреляция (PACF)
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(time_series_data, ax=ax[0])
plot_pacf(time_series_data, ax=ax[1])
plt.show()
```

**2. Проверка стационарности:**
   - **ADF Test Statistic**: -1.886425
   - **p-value**: 0.338474
   - Критические значения: 
       - 1%: -3.498910
       - 5%: -2.891516
       - 10%: -2.582760
   - P-значение больше общепринятого порогового значения 0.05, что указывает на отсутствие стационарности в данных.

**3. Автокорреляция (ACF) и частичная автокорреляция (PACF):**
   - ACF и PACF показывают, как значения временного ряда коррелируют с предыдущими значениями с учетом разных лагов.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the data
data = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')

# Grouping data by 'Год постройки' and getting the sum of 'Общая площадь'
time_series = data.groupby('Год постройки')['Общая площадь'].sum()

# Decomposition of the time series
decomposition = sm.tsa.seasonal_decompose(time_series, model='additive', period=1)

# Plotting the original time series and its decomposition into trend, seasonal, and residual components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(time_series, label='Original')
plt.legend(loc='upper left')
plt.title('Time Series Decomposition')
plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonal')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residual')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

**Выводы по Декомпозиции:**

1. **Оригинал**: Исходный ряд показывает динамику общей площади постройки домов в зависимости от года.

2. **Тренд**: Компонент тренда показывает общую тенденцию развития временного ряда. Из графика видно, что существует повышающийся тренд в общей площади постройки домов, особенно выраженный в последние годы.

3. **Сезонность**: Компонент сезонности, похоже, не имеет выраженной формы в данных (все значения равны нулю), что может быть связано с тем, что данные представлены в годовом разрезе и не содержат информации о месячных или квартальных колебаниях.

4. **Остаток**: Компонент остатка представляет собой разницу между исходными данными и предсказанными значениями, полученными на основе тренда и сезонности. Этот компонент может содержать информацию о неожиданных событиях или выбросах в данных.


```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel('i:\\!ИИ_УНТИ_МФТИ\\000\\Датасеты\\Cleaned_Data_Full_Normal.xlsx')

# Extracting the time series for total area
time_series = df.groupby("Год постройки")["Общая площадь"].sum()

# Visualizing the time series
plt.figure(figsize=(10, 5))
plt.plot(time_series.index, time_series.values, marker='o')
plt.xlabel('Year of Construction')
plt.ylabel('Total Area')
plt.title('Total Area Over the Years')
plt.grid(True)
plt.show()

# ACF and PACF plots
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(time_series, ax=ax[0])
plot_pacf(time_series, ax=ax[1])
plt.show()

# Fit ARIMA model
model = ARIMA(time_series, order=(2,1,2))
results = model.fit()

# Summary of the model
print(results.summary())

# Diagnosing the model
results.plot_diagnostics(figsize=(10, 7))
plt.tight_layout()
plt.show()

# Forecasting
forecast_steps = 5  
forecast = results.get_forecast(steps=forecast_steps)
mean_forecast = forecast.predicted_mean
conf_int = forecast.conf_int()

# Plotting the forecast along with confidence intervals
plt.figure(figsize=(10, 5))
plt.plot(time_series.index, time_series.values, label='Observed')
plt.plot(mean_forecast.index, mean_forecast.values, color='r', label='Forecast')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='pink')
plt.xlabel('Year of Construction')
plt.ylabel('Total Area')
plt.legend()
plt.title('Forecasting Total Area for Next 5 Years')
plt.grid(True)
plt.show()
```

#### ACF и PACF анализ:
